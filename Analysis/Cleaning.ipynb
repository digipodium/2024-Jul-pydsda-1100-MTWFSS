```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('laptops.csv')
df
```

```python
df.info()
```

```python
def clean_price_column(price):
    if isinstance(price,str): # isinstance() function returns True if the specified object is of the specified type, otherwise False.
        price = price.replace(',','') # replace() method replaces a specified phrase with another specified phrase.
        if price.isnumeric(): # isnumeric() method returns True if all the characters are numeric (0-9), otherwise False.
            price = int(price)
            return price
        if len(price) == 0:
            return np.nan
```

```python
clean_price_column('39,990')
```

```python
df['Cleaned_Price'] = df['price'].apply(clean_price_column)
df
```

```python
df.drop(columns='price',inplace=True)
df
```

```python
def clean_rating_column(rating):
    if isinstance(rating,str):
        rating = rating.split()[0]
        return float(rating)
    return rating
```

```python
df['Cleaned_Rating'] = df['rating'].apply(clean_rating_column)
df
```

```python
df.drop(columns='rating',inplace=True)
df
```
